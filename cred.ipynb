# Credit Card customer usage data  
  
Data Source: https://www.kaggle.com/arjunbhasin2013/ccdata

This dataset contains the usage data of few customers of a credit card. My aim is to cluster the users based on their usage behaviour and glean insights from it.

Following is the Data Dictionary for Credit Card dataset :-

CUSTID : Identification of Credit Card holder (Categorical)  
BALANCE : Balance amount left in their account to make purchases   
BALANCEFREQUENCY : How frequently the Balance is updated, score between 0 and 1 (1 = frequently updated, 0 = not 
                    frequently updated)  
PURCHASES : Amount of purchases made from account  
ONEOFFPURCHASES : Maximum purchase amount done in one-go  
INSTALLMENTSPURCHASES : Amount of purchase done in installment  
CASHADVANCE : Cash in advance given by the user  
PURCHASESFREQUENCY : How frequently the Purchases are being made, score between 0 and 1 (1 = frequently purchased,
                        0 = not frequently purchased)  
ONEOFFPURCHASESFREQUENCY : How frequently Purchases are happening in one-go (1 = frequently purchased, 
                            0 = not frequently purchased)  
PURCHASESINSTALLMENTSFREQUENCY : How frequently purchases in installments are being done (1 = frequently done, 
                                    0 = not frequently done)  
CASHADVANCEFREQUENCY : How frequently the cash in advance being paid  
CASHADVANCETRX : Number of Transactions made with "Cash in Advanced"  
PURCHASESTRX : Numbe of purchase transactions made  
CREDITLIMIT : Limit of Credit Card for user  
PAYMENTS : Amount of Payment done by user  
MINIMUM_PAYMENTS : Minimum amount of payments made by user  
PRCFULLPAYMENT : Percent of full payment paid by user  
TENURE : Tenure of credit card service for user

### Dealing With outliers
Since there are a large number of rows with outliers, instead of removing the rows i replaced the top 10 percentile of values with some value between the 90th and 91st percentile. After doing so, lets make the boxplots again.

```python
#this function kills outliers
def outl(c, r1, r2):
    c = np.where(c>r1, random.randint(r1, r2), c)
    return c

#removing outliers for each column
data["BALANCE"] = outl(data["BALANCE"], int(data['BALANCE'].quantile(0.90)), int(data['BALANCE'].quantile(0.92)))
data["PURCHASES"] = outl(data["PURCHASES"], int(data['PURCHASES'].quantile(0.90)), 
                         int(data['PURCHASES'].quantile(0.91)))
data["ONEOFF_PURCHASES"] = outl(data["ONEOFF_PURCHASES"], int(data['ONEOFF_PURCHASES'].quantile(0.90)),
                                int(data['ONEOFF_PURCHASES'].quantile(0.91)))
data["INSTALLMENTS_PURCHASES"] = outl(data["INSTALLMENTS_PURCHASES"], int(data['INSTALLMENTS_PURCHASES'].quantile(0.90)),
                                      int(data['INSTALLMENTS_PURCHASES'].quantile(0.92)))
data["CASH_ADVANCE"] = outl(data["CASH_ADVANCE"], int(data['CASH_ADVANCE'].quantile(0.90)), 
                            int(data['CASH_ADVANCE'].quantile(0.91)))
data["PAYMENTS"] = outl(data["PAYMENTS"], int(data['PAYMENTS'].quantile(0.90)), int(data['PAYMENTS'].quantile(0.91)))
data["MINIMUM_PAYMENTS"] = outl(data["MINIMUM_PAYMENTS"], int(data['MINIMUM_PAYMENTS'].quantile(0.90)),
                               int(data['MINIMUM_PAYMENTS'].quantile(0.91)))
```

```python
cols=['BALANCE', 'PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'PAYMENTS', 'MINIMUM_PAYMENTS']
plt.style.use('ggplot')
plt.figure(figsize=(16, 7))
sns.boxplot(x="value", y="variable", data=pd.melt(data[cols]), orient='h')
```

**Observations:** As we can see most of the extreme outliers have been removed

## Missing Values
As we can see MINIMUM_PAYMENTS and CREDIT_LIMIT have missing values. Since CREDIT_LIMIT has a high number of missing values lets use regression to predict and impute the missing values for it. For CREDIT LIMIT lets just use the median value.


```python
data.isnull().sum().sort_values(ascending=False).head(2)
```

```python
#replace missing value in credit_limit with median

data['CREDIT_LIMIT'].fillna((data['CREDIT_LIMIT'].median()), inplace=True)
```

```python
#delete customer id row

del data['CUST_ID']
```

```python
#split the data into train and test where the values of minimum payments are missing

y_train = data[pd.notnull(data['MINIMUM_PAYMENTS'])]['MINIMUM_PAYMENTS']
X_train = data[pd.notnull(data['MINIMUM_PAYMENTS'])]
X_test = data[pd.isnull(data['MINIMUM_PAYMENTS'])]
del X_train['MINIMUM_PAYMENTS']
del X_test['MINIMUM_PAYMENTS']
```

```python
#Train the model

from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)

#Get the predictions

y_pred = lm.predict(X_test)
```

```python
#Combine the train and test datasets to get back the original dataframe with no more missing values

X_test['MINIMUM_PAYMENTS'] = abs(y_pred)
X_train['MINIMUM_PAYMENTS'] = y_train
dat2 = X_train.append(X_test, verify_integrity=True)
dat2 = dat2.sort_index()
```

```python
dat2.isnull().sum().sort_values(ascending=False).head(2)
```

No more missing values

## Normalizing the data
Since outliers are still present in the data, we will use Robust Scaler. It is not affected by outliers since it uses Inter Quartile Range to normalize the data.

```python
from sklearn.preprocessing import RobustScaler
transformer = RobustScaler().fit(dat2)
dat2 = transformer.transform(dat2)
```

## Elbow Plot and Clustering
In order to find the optimum value for k lets use elbow method which plots sum of squared error vs different values of k. Once the data has been clustered the labels are added to the original data frame in order to visualise the clusters.

```python
from sklearn.cluster import KMeans
sse = []
kv = range(1,15)
for k in kv:
    kmc = KMeans(n_clusters=k)
    kmc.fit(dat2)
    sse.append(kmc.inertia_)
```

```python
plt.plot(kv, sse, marker='X')
```

k=6 is a good number because the drop in SSE after 6 is insignificant

```python
kmc= KMeans(6)
kmc.fit(dat2)
labels=kmc.labels_
data['CLUSTER'] = labels
```

## Visualizing the clusters
In order to visualise the 6 clusters lets plot it against the different numerical features. To get a clear image, the plots use the averages of the numerical features, averaged after grouping by clusters.

### Cash volume features
Here lets compare the clusters using cash volume features, like balance, purchases, credit limit etc.

```python
cols=['BALANCE', 'CREDIT_LIMIT', 'PAYMENTS', 'PURCHASES', 'MINIMUM_PAYMENTS', 'ONEOFF_PURCHASES',
      'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'CLUSTER']
dat3 = data[cols]
dat3=dat3.groupby('CLUSTER').mean().reset_index(drop=False)
```

```python
plt.figure(figsize=(10,10))
plt.style.use('ggplot')
palette = sns.color_palette("bright", 6)
ax = sns.scatterplot(x="component1", y="component2", hue = "cluster", data = pcdf,
                     palette =palette)
```
